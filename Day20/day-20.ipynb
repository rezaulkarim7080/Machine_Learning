```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('weight-height.csv')
df.sample(5)
```

```python
df.shape
```

## Descibe Height


```python
df['Height'].describe()
```

## Show plot


```python
sns.distplot(df['Height'])
```

```python

sns.boxplot(df['Height'])
```

### Calculate Upper Limit and lower Limit


```python
upper_limit = df['Height'].quantile(0.99)
lower_limit = df['Height'].quantile(0.01)

print("Upper limit ", upper_limit)
print("Lower limit ", lower_limit)
```

## Set Range : if criteria is not full fill


```python
df[(df['Height'] >= 74.78) | (df['Height'] <= 58.13)]
```

## Define new data framewith limit

##### (Trimming through percentile method)


```python
new_df = df[(df['Height'] <= 74.78) & (df['Height'] >= 58.13)]
new_df
```

```python
new_df.shape
```

```python
new_df['Height'].describe()
```

## Compare Before trimming and after trimming


```python
plt.figure(figsize=(16,8))

# df before trimming
plt.subplot(2,2,1)
sns.distplot(df['Height'])
plt.subplot(2,2,2)
sns.boxplot(df['Height'])

# new_df (after trimming)

plt.subplot(2,2,3)
sns.distplot(new_df['Height'])
plt.subplot(2,2,4)
sns.boxplot(new_df['Height'])
```
